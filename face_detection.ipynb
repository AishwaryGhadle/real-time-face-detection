```python
# Importing required library
import cv2

# ---------------------------------------------------------
# Step 1: Load the pre-trained face detection model
# ---------------------------------------------------------
# OpenCV provides Haarcascade files which are already trained
# to detect faces. So we don't need any dataset.

face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

# ---------------------------------------------------------
# Step 2: Start Webcam Capture
# ---------------------------------------------------------
# 0 means default webcam of laptop

video = cv2.VideoCapture(0)

print("Press 'Q' to exit the program...")

# ---------------------------------------------------------
# Step 3: Run continuous loop to read webcam frames
# ---------------------------------------------------------
while True:

    # Read frame from webcam
    ret, frame = video.read()

    # If camera not working, stop
    if not ret:
        print("Failed to grab frame")
        break

    # -----------------------------------------------------
    # Step 4: Convert image to grayscale
    # -----------------------------------------------------
    # Face detection works faster on grayscale images
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # -----------------------------------------------------
    # Step 5: Detect Faces
    # -----------------------------------------------------
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.3,   # controls detection accuracy
        minNeighbors=5     # reduces false detection
    )

    # -----------------------------------------------------
    # Step 6: Draw rectangle around detected faces
    # -----------------------------------------------------
    for (x, y, w, h) in faces:
        cv2.rectangle(
            frame,
            (x, y),
            (x + w, y + h),
            (0, 255, 0),  # Green color
            2
        )

        # Optional label
        cv2.putText(
            frame,
            "Face Detected",
            (x, y - 10),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (0, 255, 0),
            2
        )

    # -----------------------------------------------------
    # Step 7: Show Output Window
    # -----------------------------------------------------
    cv2.imshow("Real-Time Face Detection", frame)

    # Press Q to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# ---------------------------------------------------------
# Step 8: Release Camera and Close Windows
# ---------------------------------------------------------
video.release()
cv2.destroyAllWindows()
```
